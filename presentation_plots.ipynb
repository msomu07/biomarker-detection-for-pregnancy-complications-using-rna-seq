```python
bp_data = pd.DataFrame(t1_data[14586])
bp_data["disease"] = t1_diseases["disease"].values
bp_prex_data = bp_data.loc[bp_data['disease'] == "Preeclampsia"]
bp_gd_data = bp_data.loc[bp_data['disease'] == "Gestational-Diabetes"]
bp_ch_data = bp_data.loc[bp_data['disease'] == "Chronic-Hypertension"]
bp_norm_data = bp_data.loc[bp_data['disease'] == "Normal-Pregnancy"]
```

```python
data = [bp_prex_data[14586], bp_gd_data[14586], bp_ch_data[14586], bp_norm_data[14586]]

fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xticklabels(['preeclampsia', 'gestational diabetes', 'chronic hypertension','normal pregnancy'])
plt.xticks(rotation = 45)
bp = ax.boxplot(data)
plt.xlabel("disease")
plt.ylabel("read counts")
plt.title("expression of " + "FHDC1" + " in the first trimester")
plt.show()

## this box plot allows us to understand how the expression of this particular transcript may be 
## altered in patients with pregnancy complications

## i produced similar plots for many different transcripts 
```
