```python
import pandas as pd
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
normal_pregnancy_files = glob.glob("Normal-Pregnancy/RNA-seq/*.txt")
```

```python
for file in normal_pregnancy_files:
    print(file)
```

```python
name_of_NP_files = normal_pregnancy_files[0].split("/")[2].split("_")[0]
print(name_of_NP_files)
NP_df = pd.read_csv(normal_pregnancy_files[0], delimiter="\t", header = None)
NP_df.columns = ["transcript", name_of_NP_files]

for file in normal_pregnancy_files[1:]: 
    name_of_NP_files = file.split("/")[2].split("_")[0]
#     print(name_of_NP_files)
    NP_df2 = pd.read_csv(file, delimiter="\t", header = None)
    NP_df2.columns = ["transcript", name_of_NP_files]
    NP_df = pd.concat([NP_df, NP_df2[name_of_NP_files]], axis = 1)
```

```python
NP_df
```

```python
NP_df = NP_df.T ## switching the x and y axis
NP_df.drop(columns = [0, 1, 2, 3], inplace = True) ## dropping the unecessary columnns
NP_df = NP_df.T ## switching the x and y axis
NP_df.drop(columns= "transcript", axis=1, inplace = True)
NP_df = NP_df.T
NP_df
```

```python
NP_df.to_csv("normal_pregnancy_df.txt", index=False, sep="\t")
```
